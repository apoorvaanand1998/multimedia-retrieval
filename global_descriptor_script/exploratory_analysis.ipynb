```python
numeric_df = df[['Surface Area', 'Compactness',
       '3D Rectangularity', 'Diameter', 'Convexity', 'Eccentricity', 'A3',
       'D1', 'D2', 'D3', 'D4']]
numeric_df
```

```python
# split object dtypes from numeric_df
object_df = numeric_df.select_dtypes(include=['object']).copy()
object_df.head()
```

```python
# convert all columns in object_df to  numpy.ndarray
import numpy as np
# remove square brackets from object_df['A3']
"""object_df['A3'] = object_df['A3'].str.replace('[', '')
test = np.fromstring(object_df['A3'].values[0], sep=' ')
test"""
# now do this for all columns in object_df
copy_object_df = object_df.copy()
for col in copy_object_df.columns:
    copy_object_df[col] = copy_object_df[col].astype(str)
    copy_object_df[col] = copy_object_df[col].str.replace('[', '')
    copy_object_df[col] = copy_object_df[col].str.replace(']', '')
    
copy_object_df = copy_object_df.map(lambda x: np.fromstring(x, sep=' '))
```

```python
import pandas as pd

# Sample DataFrame structure with arrays in each cell
# Assuming your DataFrame is named 'copy_object_df'
original_hist_df = pd.read_csv('')
# Initialize an empty list to store expanded DataFrames
expanded_dfs = []

# Iterate through each column
for col in original_hist_df.columns:
    # Create a DataFrame where each array element becomes a separate column
    expanded_col_df = pd.DataFrame(copy_object_df[col].tolist(), index=copy_object_df.index)
    
    # Rename columns to include the original column name as a prefix
    expanded_col_df.columns = [f"{col}_{i}" for i in range(expanded_col_df.shape[1])]
    
    # Append the expanded DataFrame to the list
    expanded_dfs.append(expanded_col_df)

# Concatenate all expanded columns along with the original index
expanded_df = pd.concat(expanded_dfs, axis=1)
```

```python
# add back the Path column to expanded_df
expanded_df['Path'] = df['Path']
```

```python
file_names = global_desc_zscore['file']

class_names = []
for file in file_names:
    class_name = file.split('\\')[1]
    class_names.append(class_name)

class_names
```

```python
global_desc_zscore_values = global_desc_zscore.drop(columns=['file']).values
toy_index = 1050
toy_query_feature_vector = global_desc_zscore_values[toy_index]
toy_query_feature_vector
```

```python
class_names[toy_index]
```

```python
from scipy.spatial import distance
distances = []
num_shapes = global_desc_zscore_values.shape[0]
for i in range(num_shapes):
    dist = distance.euclidean(toy_query_feature_vector, global_desc_zscore_values[i])
    distances.append(dist)


import numpy as np
top_5_indices = np.argsort(distances)[:5]


top_5_class_names = [class_names[i] for i in top_5_indices]
top_5_class_names
```
